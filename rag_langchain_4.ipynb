```python
from google.colab import userdata
import os

os.environ["GROQ_API_KEY"]   = userdata.get("GROQ_API_KEY")
```

```python
from langchain_core.prompts import PromptTemplate
from langchain_groq import ChatGroq

chat = ChatGroq(
    model="llama-3.1-8b-instant",
    temperature=0,
    api_key=os.environ["GROQ_API_KEY"]
)
```

```python
template = """
Rol: Sen deneyimli bir yapay zeka ve derin öğrenme uzmanısın.

Görev:
Aşağıdaki kavramı teknik ama anlaşılır şekilde açıkla:

Kavram:
{concept}

Kurallar:
- Maksimum 120 kelime
- Gereksiz tekrar yapma
- Tanım + nasıl çalışır + nerede kullanılır şeklinde açıkla
- Türkçe yaz
- Başlıklar **başlık** şeklinde yaz

Çıktı formatı:
Açıklama:
"""

prompt = PromptTemplate(
    input_variables=["concept"],
    template=template,
)

chain_one = prompt | chat

response1 = chain_one.invoke({"concept": "autoencoder"})
print(response1.content)
```

```python
second_prompt = PromptTemplate(
    input_variables=["ml_concept"],
    template="""
Aşağıdaki metni referans al ve 5 yaşındaki bir çocuğa anlatır gibi açıkla.
{ml_concept}

Kurallar:
- Maksimum 120-150 kelime
- Tekrar eden cümleler yazma
- Günlük hayattan sadece 1 örnek ver
- Basit ama mantıklı bir açıklama yap
- Direkt anlatmaya başlanmalıdır.
- metnin başında merhaba gibi karşılaşma ifadeler olmamalıdır.
- metnin başında ne anlatacağını yazmadan anlatmaya başla.
"""
)

chain_two = second_prompt | chat

response2 = chain_two.invoke({"ml_concept": response1.content})
print(response2.content)
```

```python
third_prompt = PromptTemplate(
    input_variables=["st_concept"],
    template="""
Görev:
Aşağıdaki metni basit bir dille özetle.

- Metin çok temiz. Alakasız bir şey yazılmamalıdır.
- Cevabın başında 'özet' 'kısaca' gibi kelimeler olmamalıdır.
- Özet anlaşılır ve metnin gidişatını açıklamalıdır.
- Özet Türkçe olmalıdır.

Metin:
{st_concept}
"""
)

chain_three = third_prompt | chat

summary = chain_three.invoke({
    "st_concept": response1.content
}).content

response3 = f"""
**Anlatım**
{response2.content}

**Özet**
{summary}
"""

print(response3)
```

```python
from langchain_core.runnables import RunnableLambda, RunnablePassthrough

overall_chain = (
    chain_one
    | RunnableLambda(lambda x: {"ml_concept": x.content})
    | RunnablePassthrough.assign(
        child_explanation=lambda x: chain_two.invoke({"ml_concept": x["ml_concept"]}).content
    )
    | RunnableLambda(lambda x: {
        "st_concept": x["child_explanation"],
        "anlatim": x["child_explanation"]
    })
    | RunnablePassthrough.assign(
        ozet=lambda x: chain_three.invoke({"st_concept": x["st_concept"]}).content
    )
)

response4 = overall_chain.invoke({"concept": "autoencoder"})
```

```python
print("The 1. response: ", response1.content)
```

```python
print("The 3. response: ", response3)
```

```python
print("**Anlatım**\n", response4["anlatim"])
print("\n**Özet**\n", response4["ozet"])
```

```python
from google.colab import userdata
os.environ["GOOGLE_API_KEY"] = userdata.get("GOOGLE_API_KEY")
```

```python
explanation = response3
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=300,
    chunk_overlap=50,
    separators=["\n\n", "\n", ". ", " "],
)

texts = text_splitter.create_documents([explanation])

# Boş ve çok kısa chunk'ları temizle
texts = [t for t in texts if len(t.page_content.strip()) > 30]

print(f"Toplam chunk sayısı: {len(texts)}")
for i, chunk in enumerate(texts):
    print(f"\nChunk {i+1}: '{chunk.page_content}'")
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings

embeddings = GoogleGenerativeAIEmbeddings(model="models/gemini-embedding-001")

# Tek chunk embed et
query_result = embeddings.embed_query(texts[0].page_content)
print(f"Embedding boyutu: {len(query_result)}")  # 768 çıkar (HuggingFace'de 384'tü)
print(f"İlk 5 değer: {query_result[:5]}")
```

```python
all_embeddings = embeddings.embed_documents([t.page_content for t in texts])

print(f"Toplam embed edilen chunk: {len(all_embeddings)}")
for i, (chunk, emb) in enumerate(zip(texts, all_embeddings)):
    print(f"\nChunk {i+1}: '{chunk.page_content[:50]}...'")
    print(f"  → Vektör boyutu: {len(emb)}")
```
